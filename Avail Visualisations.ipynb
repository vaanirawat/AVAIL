```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Excel file for intial data
file_path = 'Avail Grower Details.xlsx'
inital_data = pd.read_excel(file_path, sheet_name = 'Form responses 1')
inital_data.head()
```

```python
#Histogram of family size
plt.figure(figsize=(6, 4))
plt.hist(inital_data['Family Size'], bins=range(int(inital_data['Family Size'].min()), int(inital_data['Family Size'].max()) + 2), 
         align='left', color='seagreen', rwidth=0.8)  
plt.title('Family Size')
plt.xlabel('Family Size')
plt.ylabel('Frequency')
plt.xticks(range(int(inital_data['Family Size'].min()), int(inital_data['Family Size'].max()) + 1))  # Setting x-ticks for each family size
plt.show()
```

```python
#Histogram of number of family members with rights to use land
plt.figure(figsize=(6, 4)) 
plt.hist(inital_data['Number of family members with rights to use land'], bins=range(int(inital_data['Number of family members with rights to use land'].min()), int(inital_data['Number of family members with rights to use land'].max()) + 2), 
         align='left', color='mediumseagreen', rwidth=0.8)  
plt.title('Number of Family Members with Rights to Use Land')
plt.xlabel('Number of Family Members')
plt.ylabel('Frequency')
plt.xticks(range(int(inital_data['Number of family members with rights to use land'].min()), int(inital_data['Number of family members with rights to use land'].max()) + 1))  # Setting x-ticks for each unit
plt.show()
```

```python
# Histogram for Number of Children at School 
plt.figure(figsize=(6, 4))  # Smaller figure size
plt.hist(inital_data['Number of Children at School'], bins=range(int(inital_data['Number of Children at School'].min()), int(inital_data['Number of Children at School'].max()) + 2), 
         align='left', color='yellowgreen', rwidth=0.8)  # Changed color, no outline, and added space between bars
plt.title('Number of Children at School')
plt.xlabel('Number of Children')
plt.ylabel('Frequency')
plt.xticks(range(int(inital_data['Number of Children at School'].min()), int(inital_data['Number of Children at School'].max()) + 1))  # Setting x-ticks for each child count
plt.show()
```

```python
# Histogram for Average number of meals per day in households with each bar representing exactly one unit
plt.figure(figsize=(6, 4))  # Smaller figure size
plt.hist(inital_data['Average number of meals per day in households'], bins=range(int(inital_data['Average number of meals per day in households'].min()), int(inital_data['Average number of meals per day in households'].max()) + 2), 
         align='left', color='green', rwidth=0.8)  # Changed color, no outline, and added space between bars
plt.title('Average Number of Meals per Day in Households')
plt.xlabel('Average Meals per Day')
plt.ylabel('Frequency')
plt.xticks(range(int(inital_data['Average number of meals per day in households'].min()), int(inital_data['Average number of meals per day in households'].max()) + 1))  # Setting x-ticks for each meal count
plt.show()
```

```python
# Pie chart for Status of Production Systems
status_counts = inital_data['Status of Production Systems'].value_counts()
plt.figure(figsize=(6, 6))  # Smaller figure size
colors = ['green', 'yellowgreen', 'olivedrab', 'darkgreen']  # Shades of green
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Status of Production Systems')
plt.show()
```

```python
# Pie chart for Environmentally Friendly Pest and Disease-Management Methods
methods_counts = inital_data['Do you use environmentally friendly pest and disease-management methods?'].value_counts()
plt.figure(figsize=(6, 6))
colors = ['yellowgreen', 'green'] 
plt.pie(methods_counts, labels=methods_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Environmentally Friendly Pest and Disease-Management Methods')
plt.show()
```

```python
# Standard boxplot for Annual Income
plt.figure(figsize=(6, 3)) 
plt.boxplot(inital_data['Annual Income'], vert=False, patch_artist=True,
            boxprops=dict(facecolor='green'),  
            medianprops=dict(color='black'))  
plt.title('Boxplot of Annual Income')
plt.xlabel('Annual Income (UGX)')
plt.show()
```

```python
# Standard boxplot for Area of Cultivated Land
plt.figure(figsize=(6, 3))
plt.boxplot(inital_data['Area of Cultivated Land'], vert=False, patch_artist=True,
            boxprops=dict(facecolor='yellowgreen'),  
            medianprops=dict(color='black'))
plt.title('Boxplot of Area of Cultivated Land')
plt.xlabel('Area of Cultivated Land (acres)')
plt.show()
```

```python
# Standard boxplot for Size of Production Cultivated
plt.figure(figsize=(8, 4))
plt.boxplot(inital_data[' Size of Production Cultivated'], vert=False, patch_artist=True,
            boxprops=dict(facecolor='mediumseagreen'),  
            medianprops=dict(color='black'))
plt.title('Boxplot of Size of Production Cultivated')
plt.xlabel('Size of Production Cultivated (kg)')
plt.show()
```
